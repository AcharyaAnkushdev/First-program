```python
#  ploting
x=np.arange(1,11)
y=[10,15,22,79,35,43,51,67,77,90]
plt.plot(x,y,'y--o',linestyle='dashed',linewidth=1,markersize=10) # r or red
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i+1,y[i]+5,y[i])

# plt.show
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("My Plot")
addlabels(x,y)
```

```python
# bar chart or frequency chart

emp5=pd.DataFrame(emp4.groupby('region').size())
emp5.reset_index(inplace=True)
emp5.rename(columns={0:'total_patient'},inplace=True)
```

```python
emp5
```

```python
#  bar plot

plt.bar(emp5['region'],emp5['total_patient'])
plt.xlabel('region')
plt.ylabel('total_patient')
plt.title('bar_chart')
```

```python
# bar chart or frequency chart

emp6=pd.DataFrame(emp4.groupby('region').agg(avg_age=('age',np.mean)))
emp6.reset_index(inplace=True)
```

```python


plt.bar(emp6['region'],emp6['avg_age'])
plt.xlabel('region')
plt.ylabel('avg_age')
plt.title('bar_chart')
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha='center',
                bbox=dict(facecolor='lightgreen',alpha=0.8))

plt.bar(emp5['region'],emp5['total_patient'],color='r',label='total_patient')
plt.bar(emp6['region'],emp6['avg_age'],color='g',label='avg_age',width=0.4)

plt.xlabel('region')
plt.title('bar_chart')
plt.legend()


addlabels(emp5['region'],emp5['total_patient'])
addlabels(emp6['region'],round(emp6['avg_age'],2))
plt.savefig(r'C:\Users\dcc\Desktop\Dataset\bar_chart.jpg')  
```

```python
plt.figure(figsize=(10,5))
colors=['lightcoral','aquamarine','plum','linen']

plt.pie(emp5['total_patient'],labels=emp5['region'],
        colors=colors,autopct='%0.1f%%',shadow=True ,startangle=90, labeldistance=0.70, pctdistance=0.5)

cent=plt.Circle((0,0), radius=0.6, fc="white", linewidth=0)
fig=plt.gcf()
fig.gca().add_artist(cent)


# startangle= srating position from 90 degrre angle
# labeldistance = label written from center from center
# white circle
# fc= face color
# gcf= fatch the current figure
# gca= get current access instance on the current figure
# pctdistance= % distance
```
